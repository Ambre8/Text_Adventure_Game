```python
"""
Docstring: 

The game is based on the movie Escape Room, released in 2019. 
The principle is simple: people are trapped in rooms with particular themes, so that they manage to get out. 
However, there is only one left. In my version, you will be the only participant. 
There are four rooms : 
- the "Waiting Room", 
- the "Pool table Room", 
- the "Hallucination Room", 
- the "Library Room". 

No bugs have been detected.

Reference : https://stackoverflow.com/questions/9246076/how-to-print-one-character-at-a-time-on-one-line

"""

import time
import sys
import random

#use delay_print() to make a typed text effect

def delay_print(s):
    for c in s:
        sys.stdout.write(c)
        sys.stdout.flush()
        time.sleep(0.025)

#start with the introduction

def start_game(): 
    delay_print("""
Welcome to my 'Escape Game'! You thought you were having fun but every mistake will be fatal. 
You are now left to your own devices. 

The first room is called the "Waiting Room" and if you don't run in two minutes, 
you'll burn at 451 DEGREES FAHRENHEIT! 

You have to find the exit before... if you know the movie you know. 
         """)
    room_1()

#define the room_1()

def room_1():
    print("""
First, where do you want to look for the exit? 
1) under the carpet.
2) behind that Frenchie bull's painting on the wall. 
3) you go to the side for safety, you're going to die anyway. 
    """)
    
#express the options behind the three choices with a conditional function 

#define the variable choice_1.

    choice_1 = input(prompt = ">  \n")
    
    if choice_1.lower()== "1" or "Carpet" in choice_1.capitalize() or "Under" in choice_1.capitalize():
        print("""\nWhy would there be an exit under the carpet? Go look somewhere else.
              Let's start again\n""")
        input(prompt = "<Press any KEY to continue>\n")
        room_1()
    
    elif choice_1.lower() == "2" or "Painting" in choice_1.capitalize() or "Frenchie" in choice_1.capitalize():
        print("""\nThere's a door, yeahhhh. Run, or the room will go up in flames.""")
        input(prompt = "<Press any KEY to continue>\n")
        room_2() #use room_2() to go directly to the next room
    
    elif choice_1.lower()== "3" or "Safety" in choice_1.capitalize() or "Die" in choice_1.capitalize():
        print("\nYou're burned")
        fail() #use fail() to make the losing player finish before the end
    
    else:
        print("Invalid command. Please try again.")
        input(prompt = "<Press any KEY to continue>\n")
        room_1()
        
#define the room_2()

def room_2():
    print("""
Well done! Too bad, I wouldn't have had to keep coding this game. Let's continue in the "Pool Table Room". 
You noticed that you have your head upside down. You see ? The exit door is on the side but without a handle. 
Oh and the "ceiling" is not so solid so you have to hurry. 
    
    The handle is in one of the first 5 pool balls. Find out which one before you fall.  
    1) Ball 1
    2) Ball 2
    3) Ball 3
    4) Ball 4
    5) Ball 5
    """)

#define variables pool_numbers and numbers

    pool_numbers = [1,2,3,4,5]
    numbers=0
    
# looping while numbers different of pool_numbers[2]  

    while numbers != pool_numbers[2]:
# use try/except to keep offering balls to the player until he finds the right one         
        try: 
            numbers = int(input(prompt = "<Choose the number of the ball you want>\n"))
   
            if numbers== pool_numbers[0]:
                print("Wrong ball, loser!\n")
        
            elif numbers== pool_numbers[1]:
                print("Are you kidding me ? WRONG!\n")

            elif numbers == pool_numbers[2]:
                print("Oh a handle! Come on, go out of this room!\n")
                input(prompt = "<Press any KEY to continue>\n")
                room_3()
    
            elif numbers == pool_numbers[3] or numbers== pool_numbers[4] :
                print("You couldn't have done it in order. WRONG!\n")
            
            else:
                print("Invalid command. Please try again.LOSER\n") 
                input(prompt = "<Press any KEY to continue>\n")
                room_2()
                
        except ValueError:
            print("Invalid command. Please try again.\n")
            
#define the room_3

def room_3():
    print("""
Wow! You did it. But it's not over, not yet. Mouahahah!
Just a minute, you've survived this far so you can tell me your first name before I move on.\n""")
    
#ask the name of the user  

    name_player = input(prompt = "What's your first name? ")
    name_player = name_player.capitalize() #capitalize allows input to always start with a capital letter.
    
    print(f"""
Ah okay {name_player} ... Actually, I don't care, you'll die anyway.
Let's have fun and find a new name for you that I can carve on your grave.\n""")
    
    input(prompt = "<Press any KEY to have a new name>\n")
    
#random.choice() allows to extract a random variable from the list_name 

    list_name = ["Batman", "Barbie", "Winnie The Poo", "Dracula", "Mario"]
    
    new_name = random.choice(list_name)

    print(f"\n What a charming little winner's name : {new_name}.\n")

    print(f"""
Let's stop talking {new_name}, you are here to die. Here you are in the "Hallucination Room"! 
This room spins around and the light is low. You can't see well but you can see a huge safe.\n""")
    
#define the variable question_safe.

    question_safe = input(prompt = """Do you want to open it ? Yes or No ?\n""")
    
#use an conditional function with two choices : yes or no

    if  question_safe.lower() == "yes":
        print ("\n Mouhahaha, poor you! Now you are hallucinating because a poison was on the lock handle.\n") 
        print ("""\nIf you want to survive, you must find the antidote, but the room is moving. 

You have three choices:
1) You let yourself die, that's your fate. 
2) You bang on everywhere without thinking. 
3) Let's see, you can see that the paint is peeling off the walls in one place.\n""")
        
#define the variable choice_2.

        choice_2 = input(prompt = ">  \n")
    
#use nested conditional statements

        if choice_2.lower()== "1" or "die" in choice_2.lower() or "fate" in choice_2.lower():
            print("\nYou  start to see blood coming out of your orifices.\n")
            input(prompt = "<Press any KEY to continue>\n")
            fail()
    
        elif choice_2.lower()== "2" or "everywhere" in choice_2.lower() or "bang" in choice_2.lower():
            print(f"""\nBut you're a complete idiot, {new_name}. Think about it. 
Well, look, you're going to die of exhaustion anyway.""")
            input(prompt = "<Press any KEY to continue>\n")
            fail()
    
        elif choice_2.lower()== "3" or "paint" in choice_2.lower() or "peeling" in choice_2.lower():
            print(f"""\nAh, {new_name}, you see. We can be nice too. 
So you find a pill you have to take and you won't have any more hallucinations. 
A door will open to take you to the last room, lucky you.""")
            input(prompt = "<Press any KEY to continue>\n")
            room_4()
    
        else:
            print("\nInvalid command. Please try again.\n")
            input(prompt = "<Press any KEY to continue>\n")
            room_3()
    
    elif question_safe.lower() == "no":
        print ("\nWell, bye then. GAME OVER.\n")
        exit(0)
        
    else:
        print("\nInvalid command. Please try again.\n")
        input(prompt = "<Press any KEY to continue>\n")
        room_3()

#define the room_4()
        
def room_4():
    print("""\nWithout transition, welcome to the "Library Room"!
    
In a few seconds the wall of the ... ah well it has already started the back wall closes in on the exit door. 
You will end up crushed if you don't find the code to open the door. 

Yes, I know, again a door but to get from one room to another you need a door so ...\n""")

#create a list of choices to facilitate the proper functioning of the conditional function

    question_code = ["How many Deadly Sins are there?",
                     "How many rooms have you been in?",
                     "How many letters make up the word 'escape'?",
                     "Give the number's place of France in the World Cup of soccer (BEST TEAM EVER)!"]
    
# looping while question with nested conditional statements in each step

    for question in question_code: 
        if question == question_code[0]:
            code_1 = input(prompt = "How many Deadly Sins are there?\n")
            
            if code_1.lower()=='7' or code_1.lower()=='seven':
                print("\nCorrect, you start well. This is the first number of your code.\n")
            elif code_1.lower()!='7':
                print("\nWRONG, from the first number!\n")
                room_4()
            else:
                print("\nInvalid command. Please try again.\n")
                input(prompt = "<Press any KEY to continue>\n")
                room_4()
        
        elif question == question_code[1]:
            code_2 = input(prompt = "How many rooms have you been in?\n")
            
            if code_2.lower()=='4' or code_2.lower()=='four':
                print("\nCorrect, again. This is the second number of your code.\n")
            elif code_2.lower()!='4':
                print("\nWRONG ! Really? You started well.\n")
                room_4()
            else:
                print("\nInvalid command. Please try again.\n")
                input(prompt = "<Press any KEY to continue>\n")
                room_4()
        
        elif question == question_code[2]:
            code_3 = input(prompt = "How many letters make up the word 'escape'?\n")
            
            if code_3.lower()=='6'or code_3.lower()=='six':
                print("\nCorrect, again. This is the third number of your code.\n")
            elif code_3.lower()!='6':
                print("\nWRONG ! Really? You started well.\n")
                room_4()
            else:
                print("\nInvalid command. Please try again.\n")
                input(prompt = "<Press any KEY to continue>\n")
                room_4()
                      
        elif question == question_code[3]:
            code_4 = input(prompt = "Give the number of the place of France in the World Cup of soccer (BEST TEAM EVER).\n")
            
            if code_4.lower()=='1' or code_4.lower()=='one':
                print("\nCorrect, but it was obvious. This is the last number of your code.\n")
                print("\nYou can run away from the last room, you are free!\n")
                input(prompt = "<Press any KEY to continue>\n")
                room_end()# allows the player to finish the game after winning
                
#use break to stop the loop when the player has the correct answer
                break 
            elif code_4.lower()!='1':
                print("\nWRONG ! Really? You started well.\n")
                room_4()
            else:
                print("\nInvalid command. Please try again.\n")
                input(prompt = "<Press any KEY to continue>\n")
                room_4()
        else:
            print("\nInvalid command. Please try again.\n")
            input(prompt = "<Press any KEY to continue>\n")
            room_4()   

#define the room_end()

def room_end():
    print("""\nYou have finished the game. Congratulations, we didn't think you could do it. 
But as they say, the clothes don't make the man.

We guess you want some explanations, so here you go. 
We are a group of wealthy people who are bored in life. 
So we created this game to bet on the success of a player, isn't it fun? 

We had bet that you won't hold a room. Well, bravo!\n""") 

    input(prompt = "<Press any KEY to continue>\n")
    delay_print ("""But we don't want a witness so... BANG! 
<YOU DEAD>\n""")
    exit(0) #allows to stop the game
    
#define the fail() function    

def fail():
    print('\nYou have failed.')
    input(prompt = '<GAME OVER>')
    exit(0) #allows to stop the game
            
start_game()
```
